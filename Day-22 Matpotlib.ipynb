```python
import matplotlib.pyplot as plt
```

```python
plt.title("Basic Liner Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
distance=list(range(2,11,2))
time=list(range(1,6))
plt.title("Distance Vs Time")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(time,distance,'ro')
plt.grid()
plt.show()
```

```python
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.title("Velocity")
plt.xlabel("Time")
plt.ylabel("Acceleration")
plt.plot(t,v)
plt.grid()
plt.show()
```
